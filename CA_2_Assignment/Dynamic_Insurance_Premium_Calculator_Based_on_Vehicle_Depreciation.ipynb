```python
import matplotlib.pyplot as plt

# Define the Vehicle class
class Vehicle:
    def __init__(self, initial_value, depreciation_rate):
        self.initial_value = initial_value
        self.depreciation_rate = depreciation_rate

    def depreciated_value(self, years):
        """Calculate the value of the vehicle after a certain number of years."""
        return self.initial_value * ((1 - self.depreciation_rate) ** years)

# Define the PremiumCalculator class
class PremiumCalculator:
    def __init__(self, vehicle):
        self.vehicle = vehicle

    def calculate_premiums(self, years):
        """Calculate monthly, quarterly, and yearly premiums based on the vehicle's value."""
        premiums = []
        for year in range(1, years + 1):
            value = self.vehicle.depreciated_value(year)
            monthly_premium = value * 0.01  # Assume monthly premium is 1% of vehicle value
            quarterly_premium = value * 0.03  # Assume quarterly premium is 3% of vehicle value
            yearly_premium = value * 0.10  # Assume yearly premium is 10% of vehicle value
            premiums.append({
                'Year': year,
                'Monthly': monthly_premium,
                'Quarterly': quarterly_premium,
                'Yearly': yearly_premium
            })
        return premiums

# Initialize Vehicle and PremiumCalculator
initial_value = 30000  # Example initial value of the vehicle
depreciation_rate = 0.07  # Depreciation rate of 7%
years = 10  # Number of years to calculate

vehicle = Vehicle(initial_value, depreciation_rate)
premium_calculator = PremiumCalculator(vehicle)
premiums = premium_calculator.calculate_premiums(years)

# Print the results
for premium in premiums:
    print(f"Year {premium['Year']}: Monthly: ${premium['Monthly']:.2f}, Quarterly: ${premium['Quarterly']:.2f}, Yearly: ${premium['Yearly']:.2f}")
```

```python
# Generate the chart
years_list = [p['Year'] for p in premiums]
monthly_premiums = [p['Monthly'] for p in premiums]
quarterly_premiums = [p['Quarterly'] for p in premiums]
yearly_premiums = [p['Yearly'] for p in premiums]

plt.figure(figsize=(10, 6))
plt.plot(years_list, monthly_premiums, marker='o', label='Monthly Premium')
plt.plot(years_list, quarterly_premiums, marker='o', label='Quarterly Premium')
plt.plot(years_list, yearly_premiums, marker='o', label='Yearly Premium')
plt.xlabel('Years')
plt.ylabel('Premium Amount ($)')
plt.title('Insurance Premiums Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
